```python
from data import data
import config

args = config.parse()
dataset, _, _ = data.load(args)
args.dataset_dict = dataset 
print(f"\nargs keys: {args.__dict__.keys()}")

X, Y, G = dataset['features'], dataset['labels'], dataset['hypergraph']
print(f"\nDefault major args : data={args.data}, dataset={args.dataset}, model_name={args.model_name}")
print(f"Data schema: X, Y, G={type(X), type(Y), type(G)}")
print(f"X shape: {X.shape}, Y shape: {Y.shape}, G len of keys: {len(G.keys())}")
```
